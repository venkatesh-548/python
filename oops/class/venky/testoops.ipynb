```python
class company:
    c_name='tech_soft_solutions'
    c_id='TEC10112'
    c_address='hyderabad'
    c_phone=8106552306
    no_emp=0
    emp_details={}
    def __init__(self,name,age,phone,email,address,sal,exp):
        self.name=name
        self.age=age
        self.phone=phone
        self.email=email
        self.address=address
        self.sal=sal
        self.exp=exp
        self.no_of_emp()
        self.emp_id=100+self.no_emp
        self.set_emp_details(self.emp_id,self)
    @classmethod
    def no_of_emp(cls):
        cls.no_emp+=1
    @classmethod
    def set_emp_details(cls,emp_id,self):
        cls.emp_details[emp_id]=self
    def sal_hike(self):
        if self.exp==1:
            self.sal=self.sal+self.sal*10/100
            print("the updated sal after sal hike:",self.sal)
        elif self.exp==2:
            self.sal=self.sal+self.sal*15/100
            print("the updated sal after sal hike:",self.sal)
        elif self.exp==3:
            self.sal=self.sal+self.sal*20/100
            print("the updated sal after sal hike:",self.sal)

        elif self.exp>=4:
            self.sal=self.sal+self.sal*25/100
            print("the updated sal after sal hike:",self.sal)
    def display_emp_details(self):
        for i in self.emp_details:
            res=self.emp_details[i]
            print(res)
            print("*"*30)
            print("employee Name:",res.name)
            print("employee id:",res.emp_id)
            print("employee Phone:",res.phone)
            print("employee Address:",res.address)
            print("employee sal:",res.sal)
            print("employee experiance:",res.exp)
    def display_spec_emp(self):
        emp_id=int(input('enter emp id:'))
        if emp_id in self.emp_details:
            temp=self.emp_details[emp_id]
            print("*"*30)
            print("employee Name:",temp.name)
            print("employee id:",temp.emp_id)
            print("employee Phone:",temp.phone)
            print("employee Address:",temp.address)
            print("employee sal:",temp.sal)
            print("employee experiance:",temp.exp)
           
        else:
            print(f'emp id {emp_id} not present in the emp details')
    @staticmethod
    def display_company_details():
        print("*"*30)
        print(company.c_name)
        print(company.c_id)
        print(company.c_address)
        print(company.c_phone)
emp1=company('venky',25,8106552306,'venky@gmail.com','dilshuknagar',30000,2)
emp2=company('ramu',24,8106552306,'ramu@gmail.com','hyd',20000,3)
emp1.sal_hike()
emp2.sal_hike()
emp1.display_emp_details()
emp2.display_spec_emp()
emp1.display_company_details() 
    
```
